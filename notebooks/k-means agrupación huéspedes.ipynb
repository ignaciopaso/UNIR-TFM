```python
# Cargar el archivo CSV subido por el usuario
file_path = 'Reservas habitaciones hotel procesado 3.csv'

# Cargar el archivo especificando el delimitador correcto (;)
df = pd.read_csv(file_path, encoding='ISO-8859-1', delimiter=';')

# Mostrar el dataset actualizado para verificar
df.head()
```

```python
import numpy as np


# Convertir la columna de fecha en tipo datetime
df['checkin'] = pd.to_datetime(df['checkin'])
# Extraer el mes y el día del mes
df['mes'] = df['checkin'].dt.month
df['dia_mes'] = df['checkin'].dt.day
# Convertir el mes en variables cíclicas
df['checkin_sin_mes'] = np.sin(2 * np.pi * df['mes'] / 12)
df['checkin_cos_mes'] = np.cos(2 * np.pi * df['mes'] / 12)
# Convertir el día del mes en variables cíclicas
df['checkin_sin_dia_mes'] = np.sin(2 * np.pi * df['dia_mes'] / 31)
df['checkin_cos_dia_mes'] = np.cos(2 * np.pi * df['dia_mes'] / 31)

# Convertir la columna de fecha en tipo datetime
df['checkout'] = pd.to_datetime(df['checkout'])
# Extraer el mes y el día del mes
df['mes'] = df['checkout'].dt.month
df['dia_mes'] = df['checkout'].dt.day
# Convertir el mes en variables cíclicas
df['checkout_sin_mes'] = np.sin(2 * np.pi * df['mes'] / 12)
df['checkout_cos_mes'] = np.cos(2 * np.pi * df['mes'] / 12)
# Convertir el día del mes en variables cíclicas
df['checkout_sin_dia_mes'] = np.sin(2 * np.pi * df['dia_mes'] / 31)
df['checkout_cos_dia_mes'] = np.cos(2 * np.pi * df['dia_mes'] / 31)

# Convertir la columna de fecha en tipo datetime
df['reserva'] = pd.to_datetime(df['reserva'])
# Extraer el mes y el día del mes
df['mes'] = df['reserva'].dt.month
df['dia_mes'] = df['reserva'].dt.day
# Convertir el mes en variables cíclicas
df['reserva_sin_mes'] = np.sin(2 * np.pi * df['mes'] / 12)
df['reserva_cos_mes'] = np.cos(2 * np.pi * df['mes'] / 12)
# Convertir el día del mes en variables cíclicas
df['reserva_sin_dia_mes'] = np.sin(2 * np.pi * df['dia_mes'] / 31)
df['reserva_cos_dia_mes'] = np.cos(2 * np.pi * df['dia_mes'] / 31)

columns_to_drop = ['checkin', 'checkout', 'reserva']
df = df.drop(columns=columns_to_drop)

# Visualizar el resultado
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Cargar el archivo CSV subido por el usuario
file_path = 'Reservas habitaciones hotel procesado 5.csv'

# Cargar el archivo especificando el delimitador correcto (;)
df_processed = pd.read_csv(file_path, encoding='ISO-8859-1', delimiter=';')
df_processed.head()
```

```python
df_processed = df_processed.replace({',': '.'}, regex=True)
# Convertir columnas a tipo numérico
for column in df_processed.columns:
    df_processed[column] = pd.to_numeric(df_processed[column], errors='coerce')
```

```python
# Normalización de los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_processed)
```

```python
# Método del codo para determinar el número óptimo de clusters
inertia = []
for k in range(1, 7):
    kmeans = KMeans(n_clusters=k, random_state=0, n_init=10, max_iter=100)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Gráfico del método del codo
plt.plot(range(1, 7), inertia, marker='o')
plt.xlabel('Número de Clusters')
plt.ylabel('Inercia')
plt.title('Método del Codo para K-Means (Dataset Lite)')
plt.show()
```

```python
# Aplicar K-Means con el número óptimo de clústeres
k_optimo = 3  # Por ejemplo, si el método del codo sugiere 3 clusters
kmeans_optimo = KMeans(n_clusters=k_optimo, random_state=0, n_init=10, max_iter=100)
df_processed['Cluster'] = kmeans_optimo.fit_predict(X_scaled)
```
